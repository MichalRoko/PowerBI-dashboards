# Data cleaning

```python
books_data = pd.read_csv('kindle_data-v2.csv')
```

```python
books_data.head()
```

```python
books_data.info()
```

```python
#get rid of useless features
drop_cols = [col for col in books_data.columns if col.endswith('URL') or col.endswith('Url')]
books_data.drop(drop_cols,axis=1,inplace=True)
```

```python
#rename columns
books_data.rename(columns={
    'asin': 'book_ID',
    'soldBy' : 'publisher',
    'stars' : 'rating',
    'reviews':'reviews_number',
    'category_name':'genre',
    'category_id' :'genre_id'
},inplace=True)
```

```python
books_data.head()
```

```python
books_data.info()
```

```python
books_data[books_data['publishedDate'].isnull()].head(20)
```

```python
#fill missing values with "Unknown"
books_data[['publisher','author']] = books_data[['publisher','author']].fillna('Unknown')
```
